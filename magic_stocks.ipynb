```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to extract data from each page
def extract_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table')
    df = pd.read_html(str(table))[0]
    return df

# Base URL
base_url = "https://www.screener.in/screens/1635381/pe-increasing/?page="

# Loop through each page from page=1 to page=35 and extract data
dfs = []
for page_num in range(1, 137):
    url = base_url + str(page_num)
    df = extract_data(url)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
final_df1 = pd.concat(dfs, ignore_index=True)

# Display the DataFrame

```
